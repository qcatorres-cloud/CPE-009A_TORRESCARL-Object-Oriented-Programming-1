**LOOP STATEMENTS**

FOR Loop

```
#Example 1
week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
for x in week:
  print(x)
```

```
#Example 2
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  print(x)
  if x == "Thursday":
    break
```

```
#Example 3(Break Statement)
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Thursday":
    break
  print(x)
```

```
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for x in week:
  if x == "Thursday":
     break
print(x)
```

```
#Example 4(Looping through a string)
for x in "week":
  print(x)
```

The range() Function

```
#Example 1
for x in range(6):
  print(x)
```

```
for x in range(2,6):
  print(x)
```

```
for x in range(2, 6, 2):
 print(x)
```

NESTED LOOPS

```
#Example 1
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print(x, y)
```

**PYTHON WHILE LOOPS**

```
#Example 5
i = 1
while i < 6:
  print(i)
  i += 1  #i = i + 1 Assignment Operators
```

```
#break statement in while loops
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```
#continue statement in while loops
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```
i = 0
while i < 6:
  i += 1
  if i != 3:
    continue
  print(i)
```

```
#The else statement
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```
