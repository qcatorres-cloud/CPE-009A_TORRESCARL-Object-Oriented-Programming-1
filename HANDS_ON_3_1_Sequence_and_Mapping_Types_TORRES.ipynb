13. Write your observation about the immutable property and the difference of assigning numberlist2 to be equal to numberlist and the numberlist.copy() method.

In **Python**, The direct assignment (=) creates a reference, meaning both of the variables point to the exact same object in memory. Because lists are **mutable**, any change to the original is reflected in the reference.

Using **.copy()** creates a shallow copy, which is a new, independent object in memory. Once cloned, the lists are no longer linked. Additionally, because the integers inside the list are immutable, updating a value simply rebinds that list index to a new number object rather than modifying the number itself.

1. Make a program that would calculate the distance between two points given a list of coordinates. Use the distance
formula.
Coordinates_list = [(1,1), (2,3)]

```python
#Distance Formula
def calculate_distance(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    distance = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
    return distance

#Coordinates list (1,1), (2,3)
coordinates_list = [(1, 1), (2, 3)]
distance = calculate_distance(coordinates_list[0], coordinates_list[1])

print("The distance between the points (1,1) and (2,3) is:", distance)
```

2. For a given string input, replace all the words “stupid” with an asterisk * equal to the length of the string. The new
string value should be displayed with the asterisks.

```python
#Simple Word Filter
import re

def simple_word_filter(text, word="stupid"):

    pattern = re.compile(re.escape(word), re.IGNORECASE)
    return pattern.sub("*" * len(word), text)

if __name__ == "__main__":
    sample_text = "Stupid Jungler, missing a Stupid smite at 1440 HP at the Baron Pit!"

    filtered_text = simple_word_filter(sample_text)

    print(f"Original: {sample_text}")
    print(f"Filtered: {filtered_text}")
```

3. Create a simple phonebook program that can read from a list of dictionaries. The program should be able to display a
person’s name, contact, and address based from a user input which is the id of the record.

```python
#Phonebook Data
phonebookList = [
    {"id": 1, "name": "Carl Torres", "contact": "793-9085", "address": "Blk 50 Lot 54 Sorrento"},
    {"id": 2, "name": "Eli", "contact": "887-9810", "address": "Teresa Street"},
    {"id": 3, "name": "Twain", "contact": "871-9820", "address": "Jollibee Street"}
]

#Search for the phone id
def find_record_by_id(record_id):
    for record in phonebookList:
        if record["id"] == record_id:
            return record
    return None

#Example
user_input = int(input("Enter the ID of the person you want to look up: "))
result = find_record_by_id(user_input)

if result:
    print("Name:", result["name"])
    print("Contact:", result["contact"])
    print("Address:", result["address"])
else:
    print("No record found with that ID.")
```
