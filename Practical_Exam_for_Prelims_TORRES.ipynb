Practical Exam:
Develop a python scripts that:

    Calculate total marks and percentage
    Assign grade for each subject based on marks:

90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

    Determine overall result (Pass/Fail)
Performance Remarks:


Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"

```
total_marks = 0
subjects_processed = 0
overall_failed = False

#User Input
subjectcount = int(input("Enter the number of subjects:"))
for i in range(subjectcount):
    Subject = input("Enter the Subject:")
    Grade = int(input("Enter the Grade:"))

    grade_letter = ""
    remark = ""
    status = ""

    if Grade < 0 or Grade > 100:
        print(f"INVALID GRADE for {Subject}!")
        continue

    subjects_processed += 1
    total_marks += Grade
#displays the letter grade marks and status on each subject
    if Grade < 40:
        grade_letter = "F"
        status = "FAILED"
        overall_failed = True
    elif Grade >= 90:
        grade_letter = "A+"
        status = "PASSED"
    elif Grade >= 80:
        grade_letter = "A"
        status = "PASSED"
    elif Grade >= 70:
        grade_letter = "B+"
        status = "PASSED"
    elif Grade >= 60:
        grade_letter = "B"
        status = "PASSED"
    elif Grade >= 50:
        grade_letter = "C"
        status = "PASSED"
    elif Grade >= 40:
        grade_letter = "D"
        status = "PASSED"

#Performance Remarks for each subject
    if Grade >= 90:
        remark = "Outstanding"
    elif Grade >= 80:
        remark = "Excellent"
    elif Grade >= 70:
        remark = "Very Good"
    elif Grade >= 60:
        remark = "Good"
    elif Grade >= 40:
        remark = "Satisfactory"
    else:
        remark = "Needs Improvement"

#Overall Performance in each subjects
    print(f"Subject: {Subject:<15} | Grade: {grade_letter:<5} | Remark: {remark:<25} | Status: {status}")

# Calculate the overall performance results
overall_percentage = 0
overall_result = ""
overall_performance_remark = ""

if subjects_processed > 0:
    overall_percentage = (total_marks / subjects_processed)

    if overall_failed:
        overall_result = "Fail"
    else:
        overall_result = "Pass"

    if overall_percentage >= 90:
        overall_performance_remark = "Outstanding"
    elif overall_percentage >= 80:
        overall_performance_remark = "Excellent"
    elif overall_percentage >= 70:
        overall_performance_remark = "Very Good"
    elif overall_percentage >= 60:
        overall_performance_remark = "Good"
    elif overall_percentage >= 40:
        overall_performance_remark = "Satisfactory"
    else:
        overall_performance_remark = "Needs Improvement"

print("\n--- Overall Results ---")
print(f"Total Marks: {total_marks}")
print(f"Overall Percentage: {overall_percentage:.2f}%")
print(f"Overall Result: {overall_result}")
print(f"Overall Performance Remark: {overall_performance_remark}")
```
