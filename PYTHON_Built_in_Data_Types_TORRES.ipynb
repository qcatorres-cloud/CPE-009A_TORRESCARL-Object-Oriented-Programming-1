**Conditional Statements**

```
from types import TracebackType
#Boolean Conditions
x = True

if x:
  print("var x is true")
else:
  print("var x is false")

#String Conditions
x = "Hello World"

if x == "Hello World":
  print("x is Hello World")
else:
  print("x is not Hello World")

#Numerical Conditions
x = 10

if x == '10':
  print("x is a String")
elif x == 10:
  print("x is a Number")
else:
  print("x is none of the above")
```

**Loops**

```
#For Loops
for var in range (0,5,2):
  print(var)

#While Loops
var = 0
while var < 5:
  print(var)
  var += 2

#Nested Loops
x = 0
while x < 5:
  for y in range(0, x):
    print(y, end='')
  x+=1
  print()
```

**FUNCTIONS**

Python functions use the following notation: **def function_name: commands**

**NOTE**: Functions are defined with the keyword **def** before the function_name. Similar to loops, the function definition ends with a colon(:)

```
def remainder(n, m):
  while True:
    if n - m < 0:
      return n
    else:
      n = n - m
remainder(10, 4)
```
