```python
# We want to create a python code for a grading scheme

# Create a function to accept a number of activities
# Compute for the score vs total
# Return the average grade of the student in %
def termCrit(actNumber):
  rawScore = []
  totalScore = []

  for i in range(0, actNumber):
    inputScore = float(input("Score: "))
    inputTotal = float(input("Total: "))

    if inputScore > inputTotal:
      print("\nInvalid Scores.\n")
      inputScore, inputTotal = 0,0
      break
    else:
      rawScore.append(inputScore)
      totalScore.append(inputTotal)

  grade = sum(rawScore) / sum(totalScore)
  grade = grade * 100
  return grade
```

```python
# Create a function for the class standing per term
def termCS():
  print("Hands-on Activities: ")
  termHOA = termCrit(int(input("How many hands-on activities this term?")))

  print("Quizzes:")
  termQuiz = termCrit(int(input("How many quizzes this term?")))

  print("Assignment: ")
  termAss = termCrit(int(input("How many assignments?")))

  grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
  return grade
```

```python
#FINAL CODE
def get_score():
    """Get a single score and total inputs from the user."""
    score = float(input("Score: "))
    total = float(input("Total: "))
    if score > total:
        print("\nInvalid Scores.\n")
        return 0, 0
    return score, total

def termCrit(actNumber):
    """Compute the grade percentage for a set of activities."""
    raw_scores = []
    total_scores = []
    for i in range(actNumber):
        score, total = get_score()
        raw_scores.append(score)
        total_scores.append(total)
    return (sum(raw_scores) / sum(total_scores)) * 100 if sum(total_scores) > 0 else 0

def termCS():
    """Compute the Class Standing grade for the term."""
    print("\nHands-on Activities:")
    termHOA = termCrit(int(input("How many hands-on activities this term? ")))
    print("\nQuizzes:")
    termQuiz = termCrit(int(input("How many quizzes this term? ")))
    print("\nAssignments:")
    termAss = termCrit(int(input("How many assignments this term? ")))
    return (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)

def termGrade(term_name, detailed=True):
    """Compute the overall grade for a term (Exam + Class Standing) or direct input."""
    if detailed:
        print(f"\n--- {term_name} Grade Computation ---")
        examRaw = float(input("Exam Score: "))
        examTotal = float(input("Exam Total: "))
        examGrade = (examRaw / examTotal) * 100 if examTotal > 0 else 0
        CSgrade = termCS()
        grade = (0.5 * examGrade) + (0.5 * CSgrade)
        print(f"{term_name} Grade: {grade:.2f}")
    else:
        grade = float(input(f"Enter {term_name} grade: "))
    return grade

def semesterGrade(detailed=True):
    """Compute the semester grade using Prelim, Midterm, and Final."""
    print("\nPrelim Grade")
    prelimGrade = termGrade("Prelim", detailed)
    print("\nMidterm Grade")
    midtermGrade = termGrade("Midterm", detailed)
    finalMG = (1/3) * prelimGrade + (2/3) * midtermGrade
    print("\nFinal Grade")
    finalGrade = termGrade("Final", detailed)
    finalFG = (1/3) * finalMG + (2/3) * finalGrade
    return prelimGrade, midtermGrade, finalFG

def convertGrade(finalGrade):
    """Convert the percentage grade to GPA scale."""
    if finalGrade >= 94.0: return 1.0
    elif finalGrade >= 88.0: return 1.25
    elif finalGrade >= 82.0: return 1.5
    elif finalGrade >= 76.0: return 1.75
    elif finalGrade >= 70.0: return 2.0
    elif finalGrade >= 64.0: return 2.25
    elif finalGrade >= 58.0: return 2.5
    elif finalGrade >= 52.0: return 2.75
    elif finalGrade >= 46.0: return 3.0
    else: return 5.0  # Failing grade

def main():
    mode = input("Compute detailed grades (y/n)? ").lower() == "y"
    prelim, midterm, final = semesterGrade(detailed=mode)
    print("\n--- Semester Results ---")
    print(f"Prelim Grade: {prelim:.2f}")
    print(f"Midterm Grade: {midterm:.2f}")
    print(f"Final Grade (Cumulative): {final:.2f}")
    print(f"GPA Equivalent: {convertGrade(final):.2f}")

if __name__ == "__main__":
    main()
```
