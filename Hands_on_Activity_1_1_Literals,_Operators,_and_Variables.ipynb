```python
# Finding a person's BMI metric
name = input("Enter your name: ") # Declare a new variable named name
weight = float(input("Enter your weight(kg): ")) # Declare a new variable named weight and convert to float
height = float(input("Enter your height(m): ")) # Declare a new variable named height and convert to float

# BMI formula: weight / (height * height)
bmi = weight / (height ** 2)

#prints the inputs
print(f"Hello {name}")
print(f"Your weight is: {weight} kg")
print(f"Your height is: {height} m")
print(f"Your BMI is: {bmi:.2f}")
```

```python
#Temperature Calculator

#Inputs
celsius = float(input("Enter the Celsius given: "))

#Formula
Fahrenheit = celsius * (9/5) + 32

#Result
print(f"The Fahrenheit of this given Celsius is {Fahrenheit:.2f}")
```

```python
#Distance Calculator between two points

#Get coordinate inputs
x1 = float(input("Enter x1: "))
y1 = float(input("Enter y1: "))
x2 = float(input("Enter x2: "))
y2 = float(input("Enter y2: "))

#Formula
distance = ((x2 - x1)**2 + (y2 - y1)**2)**0.5

#Result
print(f"The distance between the two points is: {distance:.9f}")
```
